### Práctica de Formatos de datos: XML

El objetivo de esta práctica es crear una aplicación que gestionará información sobre los parques y jardines públicos de Madrid. Para ello se van a utilizar dos APIS diferentes ofrecidas por el portal de datos abiertos del ayuntamiento de Madrid.

En primer lugar se va a usar el catálogo "Principales parques y jardines municipales". Este catálogo ofrece una relación de los parques y zonas verdes cuya gestión corresponde al Ayuntamiento de Madrid, con datos de dirección y de localización para georeferenciarlos. El siguiente trozo de código permite recuperar el archivo xml en cuestion, y lo almacena en disco.

```python
import urllib.request

x = urllib.request.urlopen('https://datos.madrid.es/egob/catalogo/200761-0-parques-jardines.xml')
a=x.read()
a=a.decode("utf-8")
prueba=open("Catalogo.xml","w")
prueba.write(str(a))
prueba.close()
```

```python
# Ejercicio 1

from xml.etree import ElementTree
import html

def parsearDireccion(atributo):
    calle = ""
    tipo = ""
    numero= ""
    localidad = ""
    provincia = ""
    cp = "" 
    for datos in atributo:
        if datos.attrib["nombre"] == "NOMBRE-VIA":
            calle = datos.text
        elif datos.attrib["nombre"] == "CLASE-VIAL":
            tipo = datos.text
        elif datos.attrib["nombre"] == "NUM":
            numero = datos.text
        elif datos.attrib["nombre"] == "LOCALIDAD":
            localidad = datos.text
        elif datos.attrib["nombre"] == "PROVINCIA":
            provincia = datos.text
        elif datos.attrib["nombre"] == "CODIGO-POSTAL":
            cp = datos.text
        elif datos.attrib["nombre"] == "BARRIO":
            break;    
    informacion = tipo+" "+calle+", "+numero+". "+provincia+". "+localidad+" "+cp+ "\n"
    return informacion



def parsearInfoParque(atributos, tags, cabeceras):
    for atributo in atributos:
         for tag in tags:
            aux = str(atributo.attrib["nombre"])
            if (aux == tag):
                print('\033[1m',cabeceras[tag],'\033[0m', ":\n")
                if atributo.attrib["nombre"] != "LOCALIZACION":
                    print(html.unescape(atributo.text), "\n")
                else:
                    print(parsearDireccion(atributo))

def mostrarInformacion(fichero, diccionarioParques, indice, tags, cabeceras):
    contador_tag = 0
    f=open(fichero,"rt")
    arbol=ElementTree.parse(f)
    for atributos in arbol.findall("contenido/atributos"):
        if int(atributos[0].text) == diccionarioParques[indice]:
            parsearInfoParque(atributos, tags, cabeceras)
    f.close()
    
#Funcion que muestra el listado de los parques    
def mostrarListaParques(fichero, listaTags):
    diccionarioParques={}
    f=open(fichero,"rt")
    arbol=ElementTree.parse(f)
    i=1
    print("Estos son los parques encontrados en el fichero XML:")
    for parque in arbol.findall("./contenido/atributos/atributo"):
        if parque.attrib["nombre"] not in listaTags:
            listaTags.append(parque.attrib["nombre"])
        if parque.attrib["nombre"]=="ID-ENTIDAD":
            diccionarioParques[i] = int(parque.text)
        if parque.attrib["nombre"]=="NOMBRE":
            print("Parque",i, ": ", parque.text)
            i=i+1
    listaTags.remove("CONTENT-URL")
    listaTags.remove("TIPO")
    listaTags.remove("DATOSCONTACTOS")
    listaTags.remove("ID-ENTIDAD")
    print("\n")
    f.close()
    return diccionarioParques



def ejercicio1(fichero):
    listaTags = []
    diccionarioCabeceras = {"NOMBRE":"Nombre del parque", "DESCRIPCION-ENTIDAD":"Descripción del parque", "TRANSPORTE":"Transporte",
                            "DESCRIPCION":"Descripción de conservación", "ACCESIBILIDAD":"Accesibilidad", "LOCALIZACION":"Dirección",
                            "EQUIPAMIENTO":"Equipamiento", "HORARIO":"Horario"}
    diccionarioParques = mostrarListaParques(fichero, listaTags)
    print(listaTags)
    indice = -1
    while indice != 0:
        indice = input("¿De qué número de parque quiere obtener la información (0 para salir)? \t")
        if indice.isdigit():
            indice = int(indice)
            if diccionarioParques.get(indice):
                print("\n")
                mostrarInformacion(fichero, diccionarioParques, int(indice), listaTags, diccionarioCabeceras)
            elif indice != 0:
                print("Indice no válido\n")
        else:
            print("Introduce un número válido\n")
    
    print("\n", '\033[1m',"BYE!",'\033[0m', "\n")
    
ejercicio1("Catalogo.xml")
```

Como resultado debería mostrarse un listado de todos los parques que cumplen las condiciones impuestas por el usuario. Se mostrará de cada parque la misma información que en el caso anterior.  En caso de no existir un parque con las condiciones dada, se mostrará un mensaje informativo.

Observar que habrá que preguntar al usuario por cada uno de los criterios, y éste tendrá que seleccionar un valor por cada criterio o no seleccionarlo.
```python
# Ejercicio 2

from xml.etree import ElementTree
import html

#Funcion auxiliar que devuelve en el parametro lista la lista actualizada con todos los valores que toma el atributo
#Dentro de la ruta especificada para el arbol del XML
def devuelveListaValores(lista, atributo, arbol, ruta):
    for parque in arbol.findall(ruta):
        if parque.attrib["nombre"]==atributo:
            if parque.text not in lista:
                lista.append(parque.text.upper())


def pedirAccesibilidad(fichero):
    print("\n", '\033[1m',"Por accesibilidad",'\033[0m', "\n")
    iniRango = None
    finRango = None
    f=open(fichero,"rt")
    arbol=ElementTree.parse(f)
    listaAccesibilidad = []
    devuelveListaValores(listaAccesibilidad, "ACCESIBILIDAD", arbol, "./contenido/atributos/atributo")
    f.close()
    print("Los valores para el campo accesibilidad del xml son: ", listaAccesibilidad)

    cadena = "Introduzca accesibilidad (-1 para ningun valor): "
    accesibilidad = input(cadena)
    if accesibilidad.lstrip("-").isdigit() and int(accesibilidad) == -1:        
        print("No se asigna valor para este criterio de búsqueda\n")
        accesibilidad = None
    elif accesibilidad not in listaAccesibilidad:
        accesibilidad = None
        print("La accesibilidad elegida no se encuentra en la lista, no se asigna valor para este criterio de búsqueda\n")
    else:
        print("El criterio para el campo ACCESIBILIDAD es:", accesibilidad,"\n")
    return accesibilidad
            

def pedirNombre(fichero):
    print("\n", '\033[1m',"Por nombre",'\033[0m', "\n")
    f=open(fichero,"rt")
    arbol=ElementTree.parse(f)
    listaNombres = []
    devuelveListaValores(listaNombres, "NOMBRE", arbol, "./contenido/atributos/atributo")
    f.close()
    print("Los nombres encontrados dentro del xml son: ",listaNombres)
    nombre = input("Introduzca nombre del parque (-1 para ningun valor)").upper()
    if nombre.lstrip("-").isdigit() and int(nombre) == -1: #esto seria un or?
        nombre = None
        print("No se asigna valor para este criterio de búsqueda\n")
    elif nombre not in listaNombres:
        nombre = None
        print("El nombre introducido no se encuentra en la lista, no se asigna valor para este criterio de búsqueda\n")
    else:
        print("El criterio para el campo NOMBRE es:", nombre,"\n")
    return nombre
    
#Funcion para pedir el criterio de barrio que muestra la lista con los distintos barrios del XML
def pedirBarrio(fichero):
    print("\n", '\033[1m',"Por barrio",'\033[0m', "\n")
    listaBarrios = []
    f=open(fichero,"rt")
    arbol=ElementTree.parse(f)
    devuelveListaValores(listaBarrios, "BARRIO", arbol, "./contenido/atributos/atributo/atributo")
    f.close()
    print("Los barrios encontrados en el xml son: ", listaBarrios)
    barrio = input("Introduzca nombre del barrio (-1 para ningun valor)").upper()
    
    if barrio.lstrip("-").isdigit() and int(barrio) == -1:
        barrio = None
        print("No se asigna valor para este criterio de búsqueda\n")
    elif barrio not in listaBarrios:
        barrio = None
        print("El barrio introducido no se encuentra en la lista, no se asigna valor para este criterio de búsqueda\n")
    else:
        print("El criterio para el campo BARRIO es:", barrio, "\n")
    return barrio
    

#Funcion para pedir el criterio de distrito que muestra la lista con los distintos distritos del XML
def pedirDistrito(fichero):
    print("\n", '\033[1m',"Por distrito",'\033[0m', "\n")
    listaDistritos = []
    f=open(fichero,"rt")
    arbol=ElementTree.parse(f)
    devuelveListaValores(listaDistritos, "DISTRITO", arbol, "./contenido/atributos/atributo/atributo")
    f.close()
    print("Los distritos encontrados en el xml son: ", listaDistritos)
    distrito = input("Introduzca nombre del distrito (-1 para ningun valor)").upper()
    if distrito.lstrip("-").isdigit() and int(distrito)==-1:
        distrito = None
        print("No se asigna valor para este criterio de búsqueda\n")
    elif distrito not in listaDistritos:
        distrito = None
        print("El distrito introducido no se encuentra en la lista, no se asigna valor para este criterio de búsqueda\n")
    else:
        print("El criterio para el campo DISTRITO es:", distrito,"\n")
    return distrito


#Funcion para parsear la direccion de un parque
def parsearDireccion(atributo):
    calle = ""
    tipo = ""
    numero= ""
    localidad = ""
    provincia = ""
    cp = "" 
    for datos in atributo:
        if datos.attrib["nombre"] == "NOMBRE-VIA":
            calle = datos.text
        elif datos.attrib["nombre"] == "CLASE-VIAL":
            tipo = datos.text
        elif datos.attrib["nombre"] == "NUM":
            numero = datos.text
        elif datos.attrib["nombre"] == "LOCALIDAD":
            localidad = datos.text
        elif datos.attrib["nombre"] == "PROVINCIA":
            provincia = datos.text
        elif datos.attrib["nombre"] == "CODIGO-POSTAL":
            cp = datos.text
        elif datos.attrib["nombre"] == "BARRIO":
            break;    
    informacion = tipo+" "+calle+", "+numero+". "+provincia+". "+localidad+" "+cp+ "\n"
    return informacion


#Funcion para escribir la informacion de un parque
def escribirInformacionParque(parque):
    cabeceras = {"NOMBRE":"Nombre del parque", "DESCRIPCION-ENTIDAD":"Descripción del parque", "TRANSPORTE":"Transporte",
                            "DESCRIPCION":"Descripción de conservación", "ACCESIBILIDAD":"Accesibilidad", "LOCALIZACION":"Dirección",
                            "EQUIPAMIENTO":"Equipamiento", "HORARIO":"Horario"}
    tags = ['NOMBRE', 'DESCRIPCION-ENTIDAD', 'TRANSPORTE', 'DESCRIPCION', 'ACCESIBILIDAD', 'LOCALIZACION', 'EQUIPAMIENTO', 'HORARIO']
    
    print("\t  ------------------------------------ \n")
    for atributo in parque:
         for tag in tags:
            aux = str(atributo.attrib["nombre"])
            if (aux == tag):
                print('\033[1m',cabeceras[tag],'\033[0m', ":\n")
                if atributo.attrib["nombre"] != "LOCALIZACION":
                    print(html.unescape(atributo.text), "\n")
                else:
                    print(parsearDireccion(atributo))

#Funcion que devuelve si una localizacion cumple el criterio de barrio
def compruebaBarrio(localizacion, barrio):
    
    for datos in localizacion:
        if datos.attrib["nombre"] == "BARRIO":
            if (barrio == datos.text):
                return True
            else:
                return False
                
#Funcion que devuelve si una localizacion cumple el criterios de distrito 
def compruebaDistrito(localizacion, distrito):
    
    for datos in localizacion:
        if datos.attrib["nombre"] == "DISTRITO":
            if (distrito == datos.text):
                return True
            else:
                return False

#Funcion que comprueba si el parque cumple los criterios de accesibilidad, distrito, barrio y nombre
def compruebaParque(parque, accesibilidad, distrito, barrio, nombre, algunResultado):
    
    #Variables boolean para marcar si cumple cada criterio, inicializadas a true si el criterio no se tiene en cuenta
    nombreCorrecto = (nombre==None)
    accesibilidadCorrecta = (accesibilidad==None)
    distritoCorrecto = (distrito==None)
    barrioCorrecto = (barrio==None)
    
    valido = (nombreCorrecto and accesibilidadCorrecta and distritoCorrecto and barrioCorrecto)
    
    if not valido:
        for atributo in parque:
            #Si cumple los cuatro criterios se termina la busqueda
            if (nombreCorrecto and accesibilidadCorrecta and distritoCorrecto and barrioCorrecto):
                break
            #Si el atributo actual corresponde al nombre y el criterio de nombre se tiene en cuenta, se comprueba si coinciden
            if atributo.attrib["nombre"] == "NOMBRE" and not nombreCorrecto:
                if (atributo.text == nombre):
                    nombreCorrecto = True

            #Si el atributo actual corresponde al accesibilidad y el criterio de accesibilidad se tiene en cuenta, se comprueba si coinciden
            if atributo.attrib["nombre"] == "ACCESIBILIDAD" and not accesibilidadCorrecta:
                if (atributo.text == accesibilidad):
                    accesibilidadCorrecta = True
            
            #Si el atributo actual corresponde a la localizacion y los criterios de barrio o distrito se tienen en cuenta
            #se comprueba si coinciden
            if atributo.attrib["nombre"] == "LOCALIZACION": 
                if not distritoCorrecto:
                    distritoCorrecto = compruebaDistrito(atributo,distrito)
                if not barrioCorrecto:
                    barrioCorrecto = compruebaBarrio (atributo, barrio)
        
    #Si el parque actual cumple todos los criterios, se llama a la funcion que escribe la informacion por pantalla
    if nombreCorrecto and accesibilidadCorrecta and distritoCorrecto and barrioCorrecto:
        if not parque.isEmpty():
            escribirInformacionParque(parque)
            algunResultado = True


def realizarBusqueda(fichero):
    accesibilidad = None
    nombre = None
    barrio = None
    distrito = None
    algunResultado = False
    listaAtributos = []
    
    accesibilidad = pedirAccesibilidad(fichero)
    
    distrito = pedirDistrito(fichero)
    
    barrio = pedirBarrio(fichero)
    
    nombre = pedirNombre(fichero)
    
    print(accesibilidad, distrito, barrio, nombre)
    # Si el usuario no ha introducido ningun dato no se busca ningun parque
    if nombre == None and accesibilidad == None and distrito == None and barrio == None:
        print("\n", '\033[1m',"Se debe instroducir mínimo 1 parámetro de búsqueda.",'\033[0m', "\n")
    else:
        f=open(fichero,"rt")
        arbol=ElementTree.parse(f)
        for parque in arbol.findall("contenido/atributos"):
            compruebaParque(parque, accesibilidad, distrito, barrio, nombre, algunResultado)
        f.close()
        
    if algunResultado == False:
        print("No se ha encontrado ningun resultado que coincida con tus criterios de búsqueda \n")
        
def salir():
    print("\n", '\033[1m',"Fin de la búsqueda",'\033[0m', "\n")
    return 0

def ejercicio2(fichero):
    indice = -1
    while indice != 0:
        print("\nMenu Busqueda\n1.- Realizar busqueda\n0.- Salir\n")
        indice = input('Escoge una: ')
        if indice.isdigit():
            indice = int(indice)
            if indice == 1:
                print("\n", '\033[1m',"Realizando búsqueda...",'\033[0m', "\n")
                realizarBusqueda(fichero)
            elif indice != 0:
                print("¡Opcion no valida!\n")
        else:
            print("¡Opcion no valida!\n")

    salir()
    
ejercicio2("Catalogo.xml")
```
